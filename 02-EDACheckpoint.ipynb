### National Park Service Visitor Use Statistics Dataset

This data came from the Query Builder for Public Use Statistics provided by the National Parks Service. There are a few complications that must be considered when using this dataset. One is that  this dataset is that 2025 data is still preliminary, however this is not a problem for us since our cutoff for date is 2024. Additionally, some parks were established after 2004, so we must be conscious in our analyses that not all parks have data for every year. Not all parks report statistics, as they may be newer and lack the resources to collect such data. Some parks also shut down for COVID-19 and thus visitor data may be affected by this, which must be considered. The National Parks Service has had a formalized method of record keeping since the 1980s, which aims to collect statistics across parks in a way that allows for statistical analysis, and upkeeps quality. Therefore, there are no glaring issues regarding our dataset for this planned use.

Regions for our dataset include parks in the following states (Using state abbreviations):
- Northeast: ME, WV, VA 
- Intermountain: UT, TX, CO, NM, MT, AZ, WY
- Midwest: SD, OH, MO, AR, IN, MI, ND, MN
- Pacific west: CA, OR, NV, HI, WA, AS

Additionally, we have defined seasons as:
- Winter: Dec-Feb
- Spring: Mar-May
- Summer: Jun-Aug
- Fall: Sep-Nov

The metrics of the table are provided below as defined by the NPS on their "Visitor Use Statistics Definitions Page" <a name="ref5"></a>[5](#footnote5):

A visitor hour is defined by the NPS as "The presence of one or more persons, excluding NPS personnel, in a park for continuous, intermittent, or simultaneous periods of time aggregating one hour (e.g., one person for one hour, two persons for one-half hour, etc. Although same day attendance at detached parts of the same park are considered duplicate counts and not reported as separate visits, the time involved at detached units is added together for the visitor hour estimate." and becomes a recreation visitor hour when "associated with a recreation visit." All of the data we will be using will be related to recreation visits (as opposed to non-recreation visits that include brief pass throughs of park land, workers doing their jobs, and research activities).

A recreation visit is defined by the NPS as "The entry of a person onto lands or waters administered by the NPS except for non-reportable and non-recreation visits. Funeral parties at National Cemeteries, school groups, etc. are reportable as ‘recreation visits’ since their use aligns with the purpose for which the park was established. Visits originating on surface vehicles (trains, boats, other) and aircraft may be counted if they stop and disembark passengers on NPS administrated territory. The applicable rule is that one entrance per individual per day may be counted."


<a name="footnote5"></a>5. [^](#ref5) National Park Service. “NPS Visitor Use Statistics Definitions” *Www.nps.gov*, 21 Feb. 2024, https://www.nps.gov/subjects/socialscience/nps-visitor-use-statistics-definitions.htm


```python
## YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
## importing pandas for the data
import pandas as pd

## loading in the dataset
national_parks = pd.read_csv('data/00-raw/2000-2024 National Park Data.csv')

## Grabbing and renaming the columns that we are going to be analyzing
national_parks_cleaned = national_parks[['ParkName', 'Region', 'State', 'Year', 'Month', 'RecreationVisits', 'RecreationHours']]
national_parks_cleaned = national_parks_cleaned.rename(columns={
    "ParkName": "Park Name",
    "RecreationVisits": "Recreation Visits",
    "RecreationHours": "Recreation Hours"
})

## Changing the Visists # and Hours # into floats for calculation purpose for the Hours Per Visit Column
national_parks_cleaned["Recreation Visits"] = pd.to_numeric(
    national_parks_cleaned["Recreation Visits"].astype(str).str.replace(",", "")
)

national_parks_cleaned["Recreation Hours"] = pd.to_numeric(
    national_parks_cleaned["Recreation Hours"].astype(str).str.replace(",", ""),
)

## Adding the Hours Per Visit Column
national_parks_cleaned['Hours Per Visit'] = (national_parks_cleaned['Recreation Hours'] / national_parks_cleaned['Recreation Visits']).round(2)

## Size of Dataset
national_parks_cleaned.shape ## (14351, 8)
```

```python
## Finding if there is missing values
national_parks_cleaned.isna().sum()

## Finding where the missing value exists
national_parks_cleaned[national_parks_cleaned.isna().any(axis=1)]

# 	Park Name	Region	State	Year	Month	Recreation Visits	Recreation Hours	Hours Per Visit
# 543	Arches NP	Intermountain	UT	2020	4	0	0	NaN
# 1143	Big Bend NP	Intermountain	TX	2020	4	0	0	NaN
# 1144	Big Bend NP	Intermountain	TX	2020	5	0	0	NaN
# 1146	Big Bend NP	Intermountain	TX	2020	7	0	0	NaN
# 2043	Canyonlands NP	Intermountain	UT	2020	4	0	0	NaN
# ...	...	...	...	...	...	...	...	...
# 13094	White Sands NP	Intermountain	NM	2020	4	0	0	NaN
# 13095	White Sands NP	Intermountain	NM	2020	5	0	0	NaN
# 13694	Yellowstone NP	Intermountain	WY	2020	4	0	0	NaN
# 13994	Yosemite NP	Pacific West	CA	2020	4	0	0	NaN
# 13995	Yosemite NP	Pacific West	CA	2020	5	0	0	NaN
# 84 rows × 8 columns

# These are the rows that includes N/A data. All of the data is for the Hours Per Visits column. This is not an error instead we can see a trend
# that all of these happens during 2020 which is the covid year, so the park most likely closed due to COVID restrictions. Therefore, there is
# 0 visits and 0 hours. Since Hours Per Visit is defined by Hours/Visits so showing NaN is expected. We are going to remove these rows as it 
# doesn't provide any insights for our analysis. Since this shows policy changes and not visitor behaviors. 

## Only keeping the rows that has recreation visists that is larger than 0
national_parks_cleaned = national_parks_cleaned[national_parks_cleaned["Recreation Visits"] > 0]

## New Size
national_parks_cleaned.shape # (14264, 8)
```

```python
## Developing a IQR range to flag outliers (For the Recreation visits)
Q1 = national_parks_cleaned['Recreation Visits'].quantile(0.25)
Q3 = national_parks_cleaned['Recreation Visits'].quantile(0.75)

## Calculating the IQR
IQR = Q3 - Q1

## Constructing an upperbound
upper_bound = Q3 + 1.5 * IQR

## Flag outliers
visits_outliers = national_parks_cleaned[national_parks_cleaned['Recreation Visits'] > upper_bound]
## Showing the highest outliers
visits_outliers.sort_values('Recreation Visits', ascending = False)
```

There are 1403 rows that are above the upper bound of the visits IQR, though this isn’t caused by errors. They are statistical outliers, because we can see that parks like Acadia and Zion are really popular and gain a lot of visitors. These observations reflect real-world park usage rather than data errors. Therefore, we will be keeping these values for the analysis.


```python
## Constructing IQR for flagging outliers (For Hours Per Visits)
q1 = national_parks_cleaned['Hours Per Visit'].quantile(0.25)
q3 = national_parks_cleaned['Hours Per Visit'].quantile(0.75)
iqr = q3 - q1

## Calculating the upper bound
hpv_upper = q3 + 1.5 * iqr

## Finding the outliers
hpv_outliers = national_parks_cleaned[national_parks_cleaned['Hours Per Visit'] > hpv_upper]
## Showing the highest outliers
hpv_outliers.sort_values('Hours Per Visit', ascending = False)
```

There are 1932 outliers according to the upper bound calculation using the IQR. We suspect that with this variable there might be more suspicious outliers. By looking at the highest few Hours Per Visit outliers we can observe some extreme numeric data. These cases are actually not data errors but reflect multiple day stays (small visitor count with long stays), or remote national parks (really far away) that require a longer stay. Since they are valid data we decided that it is appropriate to keep them in the analysis.

#### Season Trends in Park Visitation

Furthermore, we decided to provide data on seasonal trends in park visitation since it reveals how park popularity shifts from month to month. We believe a line plot is best for this visualization because it can clearly track continuous trends and changes over a set period of time. By tracking these changes in the busy and quiet periods, we can better understand exactly where parks face the highest demand for things like staffing, supplies, or maintenance.

```python
monthly_visits = (
    nps.groupby('Month')['Recreation Visits']
    .mean()
    .reset_index()
)

sns.lineplot(
    data = monthly_visits,
    x = 'Month',
    y = 'Recreation Visits',
    marker = 'o'
)
plt.title('Average Recreation Visits by Month')
plt.xlabel('Month')
plt.ylabel('Average Recreation Visits')
plt.show()
```

This line plot demonstrates a clear bell-shaped trend in average recreation visits throughout the year. We can see that visits start to climb in the spring and reach its peak in July (Month 7), with an average of over 225000 visitors. This is likely due to warmer weathers and summer vacations. After July, the numbers drop steadily through fall until they hit a low point in winter, staying around 50000 visits during December (Month 12) and January (Month 1). The steep rise and fall in the graph indicates that while the park is popular, most of that activity is packed into only a few months. This suggests how the need for park resources, supplies and staffing will be most critical during summers.


#### Visitor Behavior Across National Parks

For this next part, we are creating a barplot to analyze visitor behavior, specifically the average hours spent per visit, as it provides insight on how different parks are actually utilized by the public. We chose to create a barplot since it's the most straightforward way to rank and visually compare the different categories side by side. By examining this, we can distinguish between quick, single-visit parks and longer, multi-day-visit parks, ultimately offering a complete overview of visitor behavior across national parks.

```python
park_hpv = (
    nps.groupby('Park Name')['Hours Per Visit']
    .mean()
    .reset_index()
)

top_parks = park_hpv.sort_values(
    'Hours Per Visit',
    ascending = False
).head(15)

sns.barplot(
    data = top_parks,
    x = 'Hours Per Visit',
    y = 'Park Name'
)
plt.title('Parks with Highest Average Hours Per Visit')
plt.xlabel('Average Hours Per Visit')
plt.ylabel('Park')
plt.show()
```

This barplot shows the top 15 national parks ranked based on the average duration of visitor stays. Righ away, Isle Royale National Park stands out with the highest average duration of approximately 35 hours. This great lead is consistent with its geographic location, as it is a remote island where most people go for multi-day camping trips rather than quick single-day visits. Below that, parks like Sequoia, Kings Canyon, and Big Bend National Parks also encourages long stays, keeping visitors around for about 23-26 hours. As we move further down the graph to more famous destinations like Yellowstone, Yosemite, and the Grand Canyon, the average visit drops to roughly 15-18 hours, which seems likely to be full-day trips or single overnight stays. 

#### Behavioral Quadrant: Recreation Visits and Stay Duration

Now, we are incorporating a comparison of recreation visits and how long visitors stay. We believe using a scatterplot is most appropriate because it's especially effective in discovering correlations and identifying groupings between two different numerical variables — recreation visits and hours per visit. By mapping out these two factors, we can group parks into distinct categories, which gives us a much more complete picture of how the public interacts with the national park system.

```python
park_summary = (
    nps.groupby('Park Name')[['Recreation Visits', 'Hours Per Visit']]
    .mean()
    .reset_index()
)

sns.scatterplot(
    data = park_summary,
    x = 'Recreation Visits',
    y = 'Hours Per Visit',
    alpha = 0.7
)
plt.axvline(park_summary['Recreation Visits'].median(), linestyle = 'dotted')
plt.axhline(park_summary['Hours Per Visit'].median(), linestyle = 'dotted')

plt.title('Park Behavioral Quadrant: Popularity VS Stay Duration')
plt.xlabel('Average Recreation Visits')
plt.ylabel('Average Hours Per Visit')
plt.show()
```

This scatterplot uses dotted lines to divide the national parks into four quadrants based on their overall popularity and average stay duration. Looking at the top-left section, we can spot a few parks with low visitor numbers but long hours of visiting (including the almost 35 hour outlier we mentioned earlier). On the other hand, the bottom-right section reflects highly popular parks that draw massive crowds of over 100000 visitors, but mostly for quick trips lasting under 5 hours. Interestingly, a large cluster of dots lie in the bottom-left quadrant, suggesting that many parks experience both low visitor numbers and shorter stay times. 

#### Seasonal Visits by Region

To better understand how the seasonal patterns affect the park visitation across the United States. We want to visualize the average number of recreation visits by season and region. We want to use a barplot for this visualization because it allows us to compare the visitation level across the catagories that we want at the same time.

```python
season_map = {
    12:'Winter', 1:'Winter', 2:'Winter',
    3:'Spring', 4:'Spring', 5:'Spring',
    6:'Summer', 7:'Summer', 8:'Summer',
    9:'Fall', 10:'Fall', 11:'Fall'
}

nps['Season'] = nps['Month'].map(season_map)

season_region_visits = (
    nps.groupby(['Season', 'Region'])['Recreation Visits']
    .mean()
    .reset_index()
)

sns.barplot(
    data = season_region_visits,
    x = 'Season',
    y = 'Recreation Visits',
    hue = 'Region'
)
plt.title('Average Recreation Visits by Season and Region')
plt.show()
```

This graph clearly shows the seasonal differences in recreation visitation across the four national park regions. Across all the regions, summer is the season that peaks in visitation. The intermountain region and the Northeasts has the highest average visitation counts. We can clearly see for all the regions when the season hits the summer the visitation increases by at most of double. Winter in the other hand is the lowest season for the visitation, we can see that the visitation drops by not a lot more than halve.

### Seasonal Hours Per Visit by Region

Which correlates to the recreation visits graphs. The pacific west shows the highest average hours per visit throughout all the regions with the highest during the summer. This shows that national parks such as Yosemite and Olympic, where the visitors usually tend to stay the longest. The northeast doesn't show too much change and the intermountain shows similar data. The midwest greatly increases during the summer times and has signs of decrease throughout the other seasons.
